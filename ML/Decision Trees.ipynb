# Decision Tree Classifier

```python
from sklearn import datasets, tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import pandas as pd

# Loading Datasets
iris = datasets.load_iris()
features = iris.data
target = iris.target

X = pd.DataFrame(features, columns=iris.feature_names)
Y = target

# Creating Decision Tree Classifier object
decision_tree = DecisionTreeClassifier(random_state=0)

# Training the model
model = decision_tree.fit(features, target)

# Observation
obs = [[3, 2, 3, 2]]

# Predict observation class
print(model.predict(obs))

# Probability for each class
print(model.predict_proba(obs))

# Generating the tree
tree.plot_tree(model)
plt.show()
```

# Decision Tree Regressor

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

# Loading the dataset and splitting it into testing, training data
dataset = pd.read_csv("datasets/IceCreamData.csv")
X = dataset["Temperature"].values
Y = dataset["Revenue"].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.05)

# Create a DecisionTreeRegressor object and fit it to training data
regressor = DecisionTreeRegressor()
regressor.fit(X_train.reshape(-1, 1), Y_train.reshape(-1, 1))

# Using the regressor predict Y(Revenue) using X(Temperature) from testing data
Y_pred = regressor.predict(X_test.reshape(-1, 1))

# Create a DataFrame with real and predicted values as columns
df = pd.DataFrame(
    {"Real Values": Y_test.reshape(-1), "Predicted Values": Y_pred.reshape(-1)}
)

# Scatter plot for the regression output on testing data
plt.scatter(X_test, Y_test, color="red")
plt.scatter(X_test, Y_pred, color="green")
plt.title("Decision Tree Regression")
plt.xlabel("Temperature")
plt.ylabel("Revenue")
plt.show()

# X_grid is a set of values ranging from the lowest to the highest value in X(Temperature)
# with a seperation of 0.01 between 2 consecutive values
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))

# Plot for the regressor's predictions on testing data
plt.plot(X_grid, regressor.predict(X_grid), color="black")
plt.title("Decision Tree Regression")
plt.xlabel("Temperature")
plt.ylabel("Revenue")
plt.show()
```
