# Linear Regression for randomly generated data

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = np.random.uniform(low=-10, high=10.0, size=50)
noise = np.random.normal(loc=0.0, scale=5, size=50)

x = data
y = 2 * x + 10

plt.plot(x, y, "o")
plt.show()

features = x.reshape(-1, 1)
target = y
lr = LinearRegression()
lr.fit(features, target)

print(lr.intercept_)
print(lr.coef_)
```

# Linear Regression for Iris Data set

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

iris = pd.read_csv("datasets/Iris.csv")

data = np.random.uniform(iris.PetalLengthCm)
noise = np.random.normal(iris.PetalLengthCm)

x = data
y = 4 * x + noise
plt.plot(x, y, color="green", linestyle="", marker="*", markersize="5")
plt.show()

features = x.reshape(-1, 1)
target = y
lr = LinearRegression()
lr.fit(features, target)
print(lr.intercept_)
print(lr.coef_)
```
